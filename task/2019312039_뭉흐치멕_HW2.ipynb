## (문제)
### IMDb는 아마존의 자회사인 영화 정보 사이트로, 평점이 가장 높은 Top 250 영화를 확인할 수 있습니다.
### 해당 과제에서는 BeautifulSoup, Pandas를 이용하여 이를 시대별로 정리하고자 합니다.

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

## No. 1 (1점)
### BeautifulSoup을 이용하여 다음 url에서 제목, 연도, 평점 정보를 수집하세요.

```python
# 기본세팅
url = 'https://www.imdb.com/chart/top/'
headers = {'User-Agent' : 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser', from_encoding='utf-8')
```

```python
# 결과를 받을 빈 리스트 생성
result = []
for i in range(1,251):
    movie = soup.select('#main > div > span > div > div > div.lister > table > tbody > tr:nth-child(' + str(i) + ') > td.titleColumn > a') # 영화 이름 추출
    year = soup.select('#main > div > span > div > div > div.lister > table > tbody > tr:nth-child(' + str(i) + ') > td.titleColumn > span') # 연도 추출
    rate = soup.select('#main > div > span > div > div > div.lister > table > tbody > tr:nth-child(' + str(i) + ') > td.ratingColumn.imdbRating > strong') # 평점 추출
    for z in movie: 
        movie_2 = z.text # 영화 이름 추출 한 부분에서 text만 뽑아 저장
    for k in year:
        year_2 = k.text # 연도 추출한 부분에서 text만 뽑아 저장
    for j in rate:
        rate_2 = j.text # 평점 추출한 부분에서 text만 뽑아 저장
    result.append([movie_2, year_2, rate_2]) # 최종적으로 뽑은 내용 한 리스트에 저장
```

```python
result # 저장한 리스트 확인
```

## No. 2 (2점)
### 수집한 정보들을 데이터프레임 형태로 변환하세요.
### 해당 데이터프레임에는 순위('Rank'), 제목('Title'), 연도('Year'), 평점('Rate') 열이 존재해야 합니다.
### 특히 'Rank'와 'Year'는 int, 'Rate'는 float 자료형으로 변환하세요.

```python
df = pd.DataFrame(result) # dataframe으로 변경
```

```python
df['rank'] = range(1,251) # 순위는 해당 내용에서 가지고 오지 않았지만 순서대로 되어있으니, range만들어 df에 할당
```

```python
df.columns = ['title','year','rate','rank'] # 컬럼명 지정
```

```python
for i in range(len(df)):
    df.year[i] = df.year[i].replace('(','').replace(')','') # year에 괄호가 포함되어 있어 제거 후 astype로 타입 변경
```

```python
df['rank'] = df['rank'].astype('int')
df['year'] = df['year'].astype('int')
df['rate'] = df['rate'].astype('float')
# 각 컬럼명 지정 / dict 형태로도 변경 가능
```

```python
df = df[['rank','title','year','rate']] # 보기 편하게 rank, title, year, rate순으로 df 변경
```

```python
df # 최종 확인
```

## No. 3 (2점)
### 해당 데이터를 10년 단위로 정리하겠습니다.
### 데이터가 1920년대부터 존재하므로, 1920년대부터 2020년대까지 총 11개로 나눌 수 있습니다.
### Top 250을 대상으로, 시대별로 (1) Top 250에 포함된 영화 수, (2) 평균 평점, (3) 최고 평점 영화를 출력하세요.
### 최고 평점 영화는 시대별로 가장 순위가 높은 영화를 출력하시면 됩니다.
##### ex) 출력 결과:
##### 1920년대 - 영화 수: 6 / 평균 평점: 8.13 / 최고 평점 영화: Metropolis
##### 1930년대 - 영화 수: 7 / 평균 평점: 8.17 / 최고 평점 영화: 모던 타임즈
##### .....
##### 2020년대 - 영화 수: 6 / 평균 평점: 8.13 / 최고 평점 영화: Hamilton

```python
df2 = df.copy() # 첫번쨰 df 복사
```

```python
df2['year_10'] = df['year']//10 * 10 # 년대로 변경하기 위해 10의 몫에 10을 다시 곱해 변경
```

```python
df2 # year_10 변수 확인
```

```python
result_total = []
for decade, group in df2.groupby('year_10'): # groupby로 시대별로 묶어 진행
    count = len(group) # 해당 그룹 몇개인지 확인 = 영화 수에 해당
    mean_rating = group['rate'].mean() # 평균 개산 = 평균 평점에 해당
    max_rating = group.loc[group['rate'].idxmax(), 'title'] # 그룹내 최고 평점 확인 = 최고 평점 영화에 해당
    result_total.append(f"{decade} - 영화 수: {count} / 평균 평점: {mean_rating:.2f} / 최고 평점 영화: {max_rating}") # 해당 내용 묶어 list에 추가

# 결과 출력
for line in result_total:
    print(line)
```
